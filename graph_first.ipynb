```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.petersen_graph()
plt.subplot(121)
nx.draw(G,with_labels=True,font_width='bold')
plt.subplot(122)
nx.draw_shell(G,nlist=[range(5,10),range(5)], with_labels=True, font_weidth='bold')
```

```python
options = {
    'node_color': 'red',
    'node_size': 100,
    'width': 3,
    'edge_color':'blue'
}
plt.subplot(221)

nx.draw_random(G, **options)
plt.subplot(222)

nx.draw_circular(G, **options)
plt.subplot(223)

nx.draw_spectral(G, **options)
plt.subplot(224)

nx.draw_shell(G, nlist=[range(5,10), range(5)], **options)
```
