```python
import matplotlib.pyplot as plt
import networkx as nx

G=nx.house_graph()
pos={0:(0,0),
     1:(1,0),
     2:(0,1),
     3:(1,1),
     4:(0.5,2.0)}
nx.draw_networkx_nodes(G,pos,node_size=2000,nodelist=[4],node_color='blue')
nx.draw_networkx_nodes(G,pos,node_size=3000,nodelist=[0,1,2,3])
nx.draw_networkx_edges(G,pos,alpha=0.5,width=6)
plt.savefig("hus")
plt.axis('off')
plt.show()
```
